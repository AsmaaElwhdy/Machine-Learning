#### Identify input ,output

```python
x=DATASET.iloc[:,:-1].values         #input
y=DATASET.iloc[:,-1].values          #output
pd.DataFrame(x)
```

#### Data Cleaning

```python
from sklearn.impute import SimpleImputer as si
imputer_bmi=si(missing_values=np.nan,strategy='mean')
x[:,8:9]=imputer_bmi.fit_transform(x[:,8:9])

imputer_smoking_status=si(missing_values=np.nan,strategy='most_frequent')
x[:,9:10]=imputer_smoking_status.fit_transform(x[:,9:10])
pd.DataFrame(x)
```

#### Data Encoding

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,MinMaxScaler
lb=LabelEncoder()
x[:,0]=lb.fit_transform(x[:,0])
x[:,4]=lb.fit_transform(x[:,4])
x[:,6]=lb.fit_transform(x[:,6])

from sklearn.compose import ColumnTransformer
OneHotEncoder=ColumnTransformer([('encoder',OneHotEncoder(),[5,9])],remainder='passthrough')
x=OneHotEncoder.fit_transform(x)
pd.DataFrame(x)
```

#### Data Normalization

```python
mms=MinMaxScaler()
x[:,1:]=mms.fit_transform(x[:,1:])
pd.DataFrame(x)
```

#### Data Spliting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.7,random_state=5)
```

## Machine Learning Models

### 1_Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score , classification_report
rfc=RandomForestClassifier()
```

```python
rfc=RandomForestClassifier()
rfc.fit(x_train, y_train)
y_pred=rfc.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```



### 2_KNN Model

```python
from sklearn.neighbors import NearestCentroid
KNN=NearestCentroid()
KNN.fit(x_train, y_train)
y_pred=KNN.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```


### 3_Naive_bayes

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes_model=MultinomialNB()
naive_bayes_model.fit(x_train, y_train)
y_pred=naive_bayes_model.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
